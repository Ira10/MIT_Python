## Try it

```python
# Count unique letters in the string

s = 'abca'

unique =''

for letter in s:
    
    print(letter)
    if letter not in unique:
        unique+= letter 
    # print("unique ere", unique)
print("unique letters are : ", unique, ", with count", len(unique))
```

## Exhaustive Search

```python
cube = int(input("enter no"))

for guess in range(cube+1):
    if guess**3 == abs(cube):
        print("cube root is ", guess)
        
```

## Word Problem  ~ Computational Thinking

```python
x = 0
alyssa = x
ben = x - 2
cindy = 2 * x

for i in range(11):
    print(i)
    # if alyssa * i + ben * i + cindy * i == 10:
    if  i +  (i - 2) + (2 * i) == 10:
        print("alyssa sold ", i , "candies")
        break 
    else:
        print("we don't know")
```
